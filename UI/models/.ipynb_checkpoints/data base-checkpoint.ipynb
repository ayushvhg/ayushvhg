```python
import mysql.connector
import pandas as pd

# MySQL Connection Configurations
db_config = {
    "host": "localhost",
    "user": "root",
    "password": "root",
    "database": "multi_disease_prediction"
}

# Establish MySQL Connection
try:
    conn = mysql.connector.connect(**db_config)
    cursor = conn.cursor()
    print("Connected to MySQL successfully!")
except mysql.connector.Error as err:
    print(f"Error: {err}")
    exit()

# Load CSV Data
csv_file = "disease_natural_therapy_yoga.csv"  # Change this to your actual CSV file name
df = pd.read_csv(csv_file)

# Insert Data into MySQL Table
insert_query = """
    INSERT INTO disease 
    (disease_name, description, natural_therapy, yoga_beginner, image_name_beginner, yoga_moderate, image_name_moderate, yoga_advanced, image_name_advanced) 
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)
"""

# Convert DataFrame to List of Tuples
data_to_insert = [tuple(row) for row in df.itertuples(index=False, name=None)]

try:
    cursor.executemany(insert_query, data_to_insert)
    conn.commit()
    print(f"{cursor.rowcount} records inserted successfully!")
except mysql.connector.Error as err:
    print(f"Error inserting data: {err}")
    conn.rollback()

# Close Connection
cursor.close()
conn.close()
print("MySQL connection closed.")
```
